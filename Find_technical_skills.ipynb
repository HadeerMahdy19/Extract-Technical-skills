```python
import requests
import nltk
import pandas as pd
import numpy as np
```

### steps:
- lower strings
- drop duplicates
- filter skills from puctuation and stop words




```python
data_url = 'https://raw.githubusercontent.com/HadeerMahdy19/task1/main/Raw_Skills_Dataset.csv'
```

```python
raw_df = pd.read_csv(data_url, encoding='unicode_escape')
raw_df.tail(10)
```

```python
raw_df.shape
```

### Remove punctuation and stop words

```python
nltk.download('stopwords')
```

```python
nltk.download('punkt')
```

```python
stop_words = set(nltk.corpus.stopwords.words('english'))
```

--------------------------------------------------------------------------

```python
def preprocess_data(data):
  # convert all skills to lower case 
  data_lower = data.iloc[:,0].apply(lambda s: s.lower())

  # drop duplicates
  data_lower.drop_duplicates(inplace = True)
  print(f"Data size after lowering capitals + dropping duplicates : {data_lower.shape}")

  # extract tokens from each skill
  tokenized_df = data_lower.apply(lambda s: nltk.tokenize.word_tokenize(s))

  preprocessed_df = tokenized_df.copy()

  # filter each skill from stop words and puctuations
  for i in tokenized_df.index: 
    filtered_skill = []    # store one skill tokens after cleaning

    for skill_token in tokenized_df[i]:
      if skill_token not in stop_words and skill_token.isalpha():
        filtered_skill.append(skill_token)
    preprocessed_df[i] = ' '.join(filtered_skill)  # make filtered skill tokens into one sentence

  # drop duplicates after filtering
  preprocessed_df.drop_duplicates(inplace = True)
  print(f"Data size after tokenization + dropping stop words & punct + dropping duplicates : {preprocessed_df.shape}")
 
  return preprocessed_df
```

- Process and clean our skills df 

```python
preprocessed_df = preprocess_data(raw_df)
```

- Process and clean technical skills df from emsi API

```python
technical_skills_df = pd.read_csv('https://raw.githubusercontent.com/HadeerMahdy19/task1/main/technical_skills.csv', encoding='unicode_escape')
```

```python
preprocessed_technical_skills_df = preprocess_data(technical_skills_df.drop(['Unnamed: 0'],axis=1))
```

- Find similar skills from preprocessed_df in preprocessed_technical_skills_df

```python
res = []
for s in preprocessed_df:
  if np.sum(preprocessed_technical_skills_df.str.contains(s, case=False)) >= 1:
    res.append(s)
```
